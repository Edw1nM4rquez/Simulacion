# **Personas Empleadas por año**

```python
#empleados
def graficaEmpleados(final):
    fig, ax = plt.subplots(figsize =(16, 7))
    plt.bar(final['Periodo'], final['Total'])

    try:
        plt.ticklabel_format(axis='y', style='plain')
    except AttributeError:
        print('')
    plt.title('Personas Empleadas por Año')
    plt.xlabel('Año')
    plt.ylabel('Numero Empleados')
    plt.grid(linestyle='--', linewidth=0.4)
    for index,data in enumerate(final["Total"]):
        plt.text(x=index , y =data+1 , s=f"{data}" , fontdict=dict(fontsize=8), ha='center', va='bottom')
    plt.tight_layout()
    plt.show()
    print('*** Reporte de personas empleadas por año ***')
    print('')
    return final , fig
```

 **Generar dataframe (Empleados y Desempleados) por año**

```python
  #Generara dataframe por año#
def generarDataFrameAño(init,end):
    end = end +1
    if init >= 7 and init < end and end > init and end <=22 :
      df = pd.DataFrame()
      df_desempleo = pd.DataFrame()

      columnas = ['Periodo','Indicadores', 'Total']

      final = pd.DataFrame(columns=columnas)
      final_desempleo = pd.DataFrame(columns=columnas) 

      vect_promedio=[]
      vect_año= []
      vect_indicadores = []

      vect_promedio_D=[]
      vect_año_D= []
      vect_indicadores_D = []

      for i in range(init,end):
        if i>0 and i<=9:
          año = '0'+str(i)
        else:
          año = i
        
        meces = ['ene','feb','mar','abr','may','jun','jul','ago','sep','oct','nov','dic']

        for mes in meces:
          aux = mes+'-'+str(año)
          new_data = new_poblacion[new_poblacion.Periodo == aux]
          new_data = new_data[new_data.Indicadores == 'Empleo']

          new_data_desempleo = new_poblacion[new_poblacion.Periodo == aux]
          new_data_desempleo =  new_data_desempleo[new_data_desempleo.Indicadores == 'Desempleo']

          if len(new_data.columns):
            df = pd.concat([df,new_data])
            df_desempleo = pd.concat([df, new_data_desempleo])

        main = df.Total.mean()
        vect_promedio.append(main)
        vect_año.append(str(año))
        vect_indicadores.append('Empleo')
        df = pd.DataFrame()

        main_D = df_desempleo.Total.mean()
        vect_promedio_D.append(main_D)
        vect_año_D.append(str(año))
        vect_indicadores_D.append('Desempleo')
        df_desempleo = pd.DataFrame()

      #Generara Dataframe de Empleo
      final['Periodo']= vect_año;
      final['Indicadores']= vect_indicadores;
      final['Total']= vect_promedio;

      #Generar DataFrame de Desempleo
      final_desempleo['Periodo']= vect_año_D;
      final_desempleo['Indicadores']= vect_indicadores_D;
      final_desempleo['Total']= vect_promedio_D;

      return final, final_desempleo
```


# **Sectorizacion de Empleo**

```python
csv_pie = new_poblacion

def sectorizacion(df_period):
  
  df_es_res = df_period.groupby(['Periodo', 'Indicadores'], as_index=False)['Total'].mean()
  df_es_res.rename(columns ={
    'Indicadores':'Sector'
  }, inplace = True)
  es_fig = px.pie(df_es_res, values='Total', names='Sector', title='Sectorización de Empleo')
  es_fig.show()
  return es_fig, df_es_res

def pie(init='07', end='21'):
  try:
    df_pie2 = csv_pie.copy()
    
    df_pie2['Periodo'] = df_pie2['Periodo'].str[-2:]
    
    df_pie2 = df_pie2.loc[(df_pie2['Periodo']>=init) & (df_pie2['Periodo']<=end)]
    
    es_fig, df_es_res = sectorizacion(df_pie2)
    return csv_pie, es_fig
  except:
    print("Error rango de años invalido rangos permitidos desde 07 hasta el 21")
```
